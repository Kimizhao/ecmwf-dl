# 下载最新的0.25°ECMWF公开预报数据

## 项目背景

目前，世界主要的全球数值模式有欧洲中期数值、美国全球模式、中国全球模式等。其中，欧洲中期数值天气预报中心（ECMWF）开发的欧洲中期数值，自1979年发布以来，逐步建立了在数值天气预报领域中的领先地位，受到了气象科研人员和相关行业的广泛关注。  

对寒潮、热浪、台风等大型天气，欧洲中期数值素来有“神算子”之称，世界各国的天气预报，欧洲中期数值产品都是重要的参考。  

从2022年1月25日起，ECMWF 在全球范围内的大量预测数据将公开提供。  

从2024年2月1日起，EMCWF 公开预报数据的分辨率从0.4°提升到0.25°！  

欧洲中期数值模式的数据开放，让公众更轻易的接触到气象产品，其中蕴藏着巨大的待挖掘价值。  

> 相关链接：https://openmeteo.substack.com/p/ecmwf-ifs-upgraded-to-025-resolution

+ 气压层：1000, 925, 850, 700, 500, 300, 250, 200, 50 hPa

## 导入模块

```python
from ecmwf.opendata import Client
import numpy as np
import netCDF4 as nc
import time
import datetime
import requests
from glob import glob
from tqdm import tqdm
import os
```

## 查询最新的起报时间

```python
client = Client(source="ecmwf")
# 通过latest方法可以获取当前EC数据发布的最近起报时
start_time = client.latest(type='fc')
start_time
```

## 下载ECMWF预报数据

```python
client = Client(source='ecmwf')
client.retrieve(
        date='2024-02-21',        # 预报开始的日期
        time='18',                # 预报开始的时间（一天有4个时刻，分为00、06、12、18UTC）
        type='fc',                # 数据类型，这里fc代表forecast
        stream='oper',
        param=['2t', 'sp'],       # 变量类型，这里选择了2m温度和表面气压
        step=3,                   # 预测步长，如果当前起报时为2024年02月21号18点，那么通过该方法会返回21点的数据
        target='data.grib2')      # 文件存储名称
```

## 获取未来不同预报步长下0.4°ECMWF地面和高空预报数据

### 0.4°ECMWF地面数据

```python
client = Client(source='ecmwf')

for step in np.arange(0,12,3):
    # 通过.retrieve()方法获取想要的预报数据
    client.retrieve(
        date=start_time.strftime('%Y-%m-%d'),    #预报开始的日期
        time=start_time.strftime('%H'),          #预报开始的时间
        type='fc',                               #数据类型
        stream='oper',
        param=['2t', 'sp'],                      #变量类型，这里选择了2m温度和表面气压
        step=str(step),                          #预测步长，基于起报时间往后推
        target='ecmwf_2d_{}_fc_{}_0p4.grib2'.format(start_time.strftime('%Y%m%d%H'),str(step)), #文件存储名称
        )
```

### 0.4°ECMWF高空数据

```python
client = Client(source='ecmwf')
for step in np.arange(0,12,3):
    #通过.retrieve()方法获取想要的预报数据
    client.retrieve(
        date=start_time.strftime('%Y-%m-%d'),   #预报开始的日期
        time=start_time.strftime('%H'),         #预报开始的时间
        type='fc',                              #数据类型
        stream='oper',
        param=['u', 'v'],                       #变量类型，这里选择了2m温度和表面气压
        levelist=['850','500','200'], 
        step=str(step),           #预测步长，如果当前起报时为2024年02月21号00点，那么通过该方法会返回15点的数据
        target='ecmwf_3d_{}_fc_{}_0p4.grib2'.format(start_time.strftime('%Y%m%d%H'),str(step)), #存储名称
        )
```

## 获取未来不同预报步长下0.25°ECMWF地面和高空预报数据

### 0.25°ECMWF地面数据

```python
client = Client(source='ecmwf')

for step in np.arange(0,12,3):
    #通过.retrieve()方法获取想要的预报数据
    client.retrieve(
        date=start_time.strftime('%Y-%m-%d'),    #预测开始的日期
        time=start_time.strftime('%H'),          #预测开始的时间
        type='fc',          #预报数据
        stream='oper',
        param=['2t', 'sp'], #变量类型，这里选择了温度和气压
        step=str(step),           #预测步长，如果当前起报时为2024年02月21号00点，那么通过该方法会返回15点的数据
        target='ecmwf_2d_{}_fc_{}_0p25.grib2'.format(start_time.strftime('%Y%m%d%H'),str(step)),#存储名称
        resol='0p25',
        )
```

### 0.25°ECMWF高空数据

```python
client = Client(source='ecmwf')
for step in np.arange(0,12,3):
    #通过.retrieve()方法获取想要的预报数据
    client.retrieve(
        date=start_time.strftime('%Y-%m-%d'),    #预测开始的日期
        time=start_time.strftime('%H'),          #预测开始的时间
        type='fc',          #预报数据
        stream='oper',
        param=['u', 'v'],  #变量类型，这里选择了温度和气压
        levelist=['850','500','200'], 
        step=str(step),           #预测步长，如果当前起报时为2024年02月21号00点，那么通过该方法会返回15点的数据
        target='ecmwf_3d_{}_fc_{}_0p25.grib2'.format(start_time.strftime('%Y%m%d%H'),str(step)), #存储名称
        resol='0p25',
        )
```

## grib格式转nc格式

```python
def grib2nc(filename):

    ncname = filename.replace('.grib2','.nc')
    da = nc.Dataset(ncname,'w',format='NETCDF4')
    
    grb = pg.open(filename)
    g = grb[1]
    lat, lon = g.latlons()
    da.createDimension('lat', np.size(lat,0))
    da.createDimension('lon', np.size(lat,1))
    
    # add lat and lon variables
    da.createVariable('lat','f4',('lat','lon'),zlib=True,complevel=9,fill_value=-9999)
    da.createVariable('lon','f4',('lat','lon'),zlib=True,complevel=9,fill_value=-9999)
    da.variables['lat'][:]=lat
    da.variables['lon'][:]=lon
    
    for g in grb:
    
        # extract data and attribute
        varname = g.name
        data = g.values
        unit = g.units
        level = g.level

        
        # write into nc file
        if varname=='unknown':
            continue
        if level!=0 and level!=2 and level!=10:
            varname = varname + ' ' + str(level) + 'hPa'
        
        # print(varname)
        d = da.createVariable(varname,'f4',('lat','lon'),zlib=True,complevel=9,fill_value=-9999)
        d.units = unit
        da.variables[varname][:] = data
    
    da.close()
    grb.close()
```

```python
filelist = glob('ecmwf_2d_*.grib2')
for file in tqdm(filelist):
    grib2nc(file)
```

```python
filelist = glob('ecmwf_3d_*.grib2')
for file in tqdm(filelist):
    grib2nc(file)
```
